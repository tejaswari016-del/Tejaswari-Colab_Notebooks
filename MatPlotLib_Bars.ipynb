```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,25,20,30,40]
z = [100,200,300,400,500]

plt.plot(x,y,z,marker='^',linestyle='--',color='black',linewidth=2,markersize=8,label='Sales')
plt.title("Monthly sales trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.twinx()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Genarating random data
data = np.random.normal(170,10,250)
#Create histogram
plt.hist(data,bins=90)

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50,100,150)

plt.hist(data,bins=10, color="black",edgecolor="red")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
data = [
    30,35,40,42,45,47,48,49,50,50,51,52,53,55,57,60,62,65,70,75
]
plt.figure(figsize=(8,5))
plt.hist(
    data,bins=5,color="green",
    edgecolor="red",
    alpha=0.9,density=True,
    range=(30,80)
)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram without Random Data")
plt.grid(True)
plt.tight_layout()
plt.show()
```
