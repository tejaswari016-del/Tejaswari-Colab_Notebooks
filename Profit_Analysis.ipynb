```python
def load_data(filename):
  crops = []
  file = open(filename,"r")
  for line in file:
     line = line.strip()
     if line:
        data = line.split(",")
        crop = {
              "name":data[0],
              "season":data[1],
              "area":float(data[2]),
              "yield":float(data[3]),
              "cost":float(data[4]),
              "revenue":int(data[5])
              }
        crops.append(crop)
  return crops

def display_crops(crops):
    print("\nCrop Details:")
    for crop in crops:
        print(crop)


#Finding all Kharif crops
def find_kharif_crops(crops):
    print("\nKharif Crops:")
    for crop in crops:
        if crop["season"] == "kharif":
            print(crop["name"])


#Finding all non-Kharif crops
def find_non_kharif_crops(crops):
    print("\nNon-Kharif Crops:")
    for crop in crops:
        if crop["season"] == "rabi":
            print(crop["name"])


#Calculating profit for each crop
def calculate_profit(crops):
    print("\nProfit Details:")
    for crop in crops:
        profit = crop["revenue"] - crop["cost"]
        print(crop["name"], "Profit:", profit)


#Summary statistics
def summary_statistics(crops):
  total_area = 0
  total_yield = 0

  for crop in crops:
    total_area += crop["area"]
    total_yield += crop["yield"]

  avg_yield = total_yield / len(crops)

  print("\nSummary Statistics:")
  print("Total Area:", total_area)
  print("Average Yield:", avg_yield)

# ---------------- MAIN PROGRAM ----------------

def main():
    input_file = "profit.analysis.txt"
    output_file = "analysis_report.txt"

    crops = load_data(input_file)

    display_crops(crops)
    find_kharif_crops(crops)
    find_non_kharif_crops(crops)
    calculate_profit(crops)
    summary_statistics(crops)

main()
```
