```python
#Step1: Import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score

#Step2: Create sample agricultural dataset
data = {
    "Rainfall":[450,520,390,610,480,530,410,590],
    "Temperature":[22,20,25,19,23,21,24,18],
    "Nitrogen":[0.8,1.0,0.6,1.2,0.9,1.1,0.7,1.3],
    "Fertilizer":[120,140,100,160,130,150,110,170],
    "Yield":[3.5,4.2,2.9,4.8,3.8,4.4,3.1,5.0]
}

df = pd.DataFrame(data)

#Step3: Define Features(X) and Target(y)
X = df[["Rainfall","Temperature","Nitrogen","Fertilizer"]]
y = df["Yield"]

print(X)
print(y)

#Step4:Split data into Training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
print(X_train)
print(X_test)

#Step5: Train model
model = LinearRegression()
model.fit(X_train,y_train)       #Chooses the coefficients that minimize error and uses Least squares

#step6: Make predictions:
y_pred = model.predict(X_test)

#step7:Evaluate model
mae = mean_absolute_error(y_test,y_pred)  #Avg diff b/w actual yield and predicted yield
r2 = r2_score(y_test,y_pred)              # r2 = values b/w 0 to 1

print("Mean Absolute Error:",round(mae,2))  #smaller value=better model
print("R2 Score:",round(r2,2))              #closer to 1=very good prediction

#step8:Predict yield for New Farm Conditions
new_data = np.array([[500,  21,1.0,145]])     #Rf,Temp,Nitrogen,Fertilizer
predicted_yield = model.predict(new_data)

print("Predicted yield(tons/hectare):",round(predicted_yield[0],2))




```
