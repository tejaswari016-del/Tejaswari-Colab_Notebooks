```python
#Creating dataframes from a dictionary of lists
import pandas as pd
data = {
    "Name": ["Ramesh","Suresh","Mahesh"],
    "Age":[40,38,42],
    "No of children":[1,0,3],
    "Acres of land":[6,3,1.5]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Creating dataframes from a list of lists
import pandas as pd
data = [
    ["Ramesh",40,1,6],
    ["Suresh",38,0,3],
    ["Mahesh",42,3,1.5]
]
df = pd.DataFrame(data,columns=["Name","Age","No of children","Acres of land"])
print(df)
```
