```python
a=10
print("Initial value:",a)
a+=5
print("After+=:",a)
a-=3
print("After-=:",a)
a*=2
print("After*=:",a)
a/=4
print("After/=:",a)
a//=2
print("After//=:",a)
a%=3
print("After%=:",a)
a**=2
print("After**=:",a)
```
