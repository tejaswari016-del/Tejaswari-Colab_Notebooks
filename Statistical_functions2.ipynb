```python
import numpy as np

n = int(input("Enter number of months: ")

rainfall_list = []
yield_list = []

print("\nEnter rainfall data (in mm):")
for i in range(n):
    rain = float(input(f"Month {i+1}: "))
    rainfall_list.append(rain)

print("\nEnter crop yield data (tons/hectare):")
for i in range(n):
    y = float(input(f"Month {i+1}: "))
    yield_list.append(y)

# Converting lists to numpy arrays
rainfall = np.array(rainfall_list)
yields = np.array(yield_list)

# ================= RAINFALL STATISTICS =================
print("\n===== RAINFALL STATISTICS =====")
print(f"Mean rainfall: {np.mean(rainfall):.2f} mm")
print(f"Median rainfall: {np.median(rainfall):.2f} mm")
print(f"Standard deviation: {np.std(rainfall):.2f} mm")
print(f"Variance: {np.var(rainfall):.2f} mm")
print(f"Maximum rainfall: {np.max(rainfall)} mm (Month {np.argmax(rainfall)+1})")
print(f"Minimum rainfall: {np.min(rainfall)} mm (Month {np.argmin(rainfall)+1})")
print(f"Range: {np.ptp(rainfall)} mm")

# ================= YIELD STATISTICS =================
print("\n===== YIELD STATISTICS =====")
print(f"Mean yield: {np.mean(yields):.2f} tons/ha")
print(f"Median yield: {np.median(yields):.2f} tons/ha")
print(f"Standard deviation: {np.std(yields):.2f} tons/ha")
print(f"Variance: {np.var(yields):.2f} tons/ha")

# ================= CORRELATION =================
correlation = np.corrcoef(rainfall, yields)[0, 1]

print("\n===== CORRELATION ANALYSIS =====")
print(f"Rainfall–Yield correlation: {correlation:.4f}")

if correlation > 0.7:
    print("Interpretation: Strong positive correlation")
elif correlation > 0.4:
    print("Interpretation: Moderate correlation")
else:
    print("Interpretation: Weak correlation")

# ================= CUMULATIVE RAINFALL =================
print("\n===== CUMULATIVE RAINFALL =====")
cumulative_rainfall = np.cumsum(rainfall)

for month, rain, total in zip(range(1, n+1), rainfall, cumulative_rainfall):
    print(f"Month {month}: Rainfall = {rain} mm | Cumulative = {total} mm")
```

```python
#Calculating student examination marks
import numpy as np

n = int(input("Enter number of students: "))
s = int(input("Enter number of subjects: "))
subjects = []
print("\nEnter subject names: ")
for i in range(s):
    sub = input(f"Subject {i+1}: ")
    subjects.append(sub)
all_marks = []

print("\nEnter marks for each student: ")
for i in range(n):
   print(f"\nStudent {i+1}: ")
   student_marks = []
   for j in range(s):
       mark = float(input(f"{subjects[j]}: "))
       student_marks.append(mark)
   all_marks.append(student_marks)

marks = np.array(all_marks)

#==================STUDENT-WISE ANALYSIS==================
print("\n===== STUDENT-WISE ANALYSIS =====")
for i in range(n):
   total = np.sum(marks[i])
   average = np.mean(marks[i])
   print(f"Student {i+1} ---> Total: {total:.2f}, Average: {average:.2f}")

#=================SUBJECT WISE ANALYSIS====================
print("\n==== SUBJECT-WISE ANALYSIS ====")
for j in range(s):
    print(f"\nSubject: {subjects[j]}")
    print(f"Average marks: {np.mean(marks[:,j]):.2f}")
    print(f"Highest marks: {np.max(marks[:,j])}")
    print(f"Lowest marks: {np.min(marks[:,j])}")

total_marks = np.sum(marks,axis=1)
topper = np.argmax(total_marks)+1

print("=======OVERALL TOPPER=======")
print(f"Topper of the examination: Student{topper}")
```
