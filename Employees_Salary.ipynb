```python
 import pandas as pd
 import numpy as np

 #creating sample dataframe
 df = pd.DataFrame({
     'Name': ['Bhavana','Teju','Shubha','Shreesha','Shruti','Nayana'],
     'Department': ['HR','IT','IT','Finance','HR','Finance'],
     'Salary':[50000,70000,70000,60000,52000,60000],
     'Experience':[2,5,4,6,3,6]
     })
 print(df)
```

```python
sal = df.sort_values('Salary')
print(sal)
```

```python
sal = df.sort_values('Salary',ascending=False)
print(sal)
```

```python
sal=df.sort_values(['Salary','Experience'],ascending=[False,True])
print(sal)
```

```python
sal=df.sort_values(['Salary','Experience'])
print(sal)
```

```python
des=df.sort_index(ascending=False)
print(des)
```

```python
df['rank_salary']=df['Salary'].rank()
print(df)
```

```python
df['rank_salary'] = df['Salary'].rank(ascending=False)
print(df)
```

```python
df['Dense_Rank_Salary'] = df['Salary'].rank(ascending=False,method='dense')
print(df)
```

```python
depart_name = df['Department'].value_counts()
print(depart_name)
```

```python
df = df['Department'].value_counts(normalize=True)
print(df)
```

```python
df = df.sort_values('Name')
print(df)
```
