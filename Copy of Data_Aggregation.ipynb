```python
import pandas as pd
data = {
    'Department':['HR','Sales','IT','Sales','IT','HR','Sales'],
    'Employee':['A','B','C','D','E','F','G'],
    'Salary':[50000,60000,45000,48000,70000,72000,71000],
    'Experience':[2,4,3,5,6,7,5]
}
df = pd.DataFrame(data)
print(df)
```

```python
#.sum()
df.groupby(['Department','Experience']).sum()
```

```python
df.groupby('Department')['Salary'].sum()
```

```python
df2 = df.groupby('Department').sum()
print(df2)
```

```python
#Custom aggregations
df.groupby('Department').agg({'Salary':'sum','Experience':'mean'})
```

```python
#Broadcast group mean to each row
df9 = df['Dept_Avg_Salary'] = df.groupby('Department')['Salary'].transform('mean')
print(df9)
```

```python
#Filtering groups
df1 = df.groupby('Department').filter(lambda x : len(x)>2)
print(df1)
```

```python
#Pivot tables
#Mean salary by department and experience
df.pivot_table(
    values = 'Salary',
    index = 'Department',
    columns = 'Experience',
    aggfunc = 'mean'
)
```

```python
import pandas as pd
data = {
    'Crop':['Rice','Wheat','Maize','Rice','Wheat','Maize','Rice'],
    'Farmer':['Surya','Raju','Dinesh','Lakshmi','Monika','Bhargav','Dileep'],
    'Season':['Kharif','Rabi','Kharif','Kharif','Rabi','Kharif','Rabi'],
    'Yield':[50,45,40,55,48,42,60],
    'Area':[2,1.5,2,2.5,1.8,2,3]
    }
df = pd.DataFrame(data)
print(df)
```

```python
df1 = df.groupby(['Crop','Yield']).sum()
print(df1)
```

```python
df2 = df.groupby('Crop').sum()
print(df2)
```

```python
df.groupby('Crop').agg({
    'Yield':'sum',
    'Area':'mean'
    })
```

```python
df.groupby(['Crop','Farmer']).agg({
    'Yield':'sum',
    'Area':'mean'
    })
```

```python
df4 = df["Crop_Avg_Yield"] = df.groupby('Crop')['Yield'].transform('mean')
print(df4)
```

```python
df1 = df.groupby('Crop').filter(lambda x : len(x)>2)
print(df1)
```

```python
df.pivot_table(
    values = 'Yield',
    index = 'Crop',
    columns = 'Farmer',
    aggfunc = 'sum'
)
```

```python
import pandas as pd
data = {
    'Name':['Ravi','Asha','Siddhu','Ravi','Asha'],
    'Subject':['Math','Science','Math','Science','Social'],
    'Marks':[85,90,78,82,95]
}
df = pd.DataFrame(data)
print(df)
```
