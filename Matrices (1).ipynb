```python
import numpy as np

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
C = A + B
print("Matrix addition: ")
print(C)
```

```python
C = A*B
print("Matrix multiplication: ")
print(C)
```

```python
#Transpose of a matrix
print("Transpose of a matrix: ")
print(A.T)
print(B.T)
```

```python
#determinant of a matrix
det = np.linalg.det(A)
print(det)
```

```python
#inverse of a matrix
A = np.array([[1,2],[3,4]])
inv_A = np.linalg.inv(A)
print(inv_A)
```

```python
#rank of a matrix
rank = np.linalg.matrix_rank(A)
print(rank)
```

```python
values,vectors = np.linalg.eig(A)
print(values)
print(vectors)
```

```python
#linear equations(Ax = b)
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

x = np.linalg.solve(A,B)
print(x)
```
