```python
import numpy as np

#crop yields for 4 farms over 5 seasons(in tons)
crop_yields = np.array([
    [12,13.2,14.1,13.8,15.2],
    [10,11.5,10.8,12.1,11.9],
    [15,14.8,16.2,15.5,16.8],
    [11,10.9,12.3,11.7,12.8]
])
print(' Crop yields array(tons): ')
print(crop_yields)
print(f'\nShape: {crop_yields.shape}')
print(f'Data type: {crop_yields.dtype}')
print(f'Total elements: {crop_yields.size}')
print(f'Number of dimensions: {crop_yields.ndim}')

 #Agricultural analysis operations
total_yield = np.sum(crop_yields)
avg_yield_per_farm = np.mean(crop_yields,axis=1)
avg_yield_per_season = np.mean(crop_yields,axis=0)
max_yield = np.max(crop_yields)
min_yield = np.min(crop_yields)

print(f'\nTotal production:{total_yield}tons')
print(f'Average yield per farm: {avg_yield_per_farm}tons')
print(f'Average yield per season: {avg_yield_per_season}tons')
print(f'Maximum yield: {max_yield}tons')
print(f'Minimum yield: {min_yield}tons')

 #Find best performing farm
best_farm = np.argmax(avg_yield_per_farm)

print(f'\nBest performing farm:Farm{best_farm + 1}')
print(f'Their average yield:{avg_yield_per_farm[best_farm]: 2f}tons')
```

```python
import numpy as np

#Marks of a students in 5 subjects

marks = np.array([
    [78,85,90,88,92],  #student1
    [65,70,72,68,75],   #student2
    [70,75,80,78,82],   #student3
    [88,92,95,90,94]   #student4
])
print("Students marks array: ")
print(marks)

print(f"\nShape: {marks.shape}")
print(f"Data type: {marks.dtype}")
print(f"Total elements: {marks.size}")
print(f"Number of dimensions: {marks.ndim}")

total_marks = np.sum(marks)
avg_marks_per_student = np.mean(marks,axis=1)
avg_marks_per_subject = np.mean(marks,axis=0)
max_marks = np.max(marks)
min_marks = np.min(marks)

print(f"\nTotal marks of all students: {total_marks}")
print(f"Average marks per student: {avg_marks_per_student}")
print(f"Average marks per subject: {avg_marks_per_subject}")
print(f"Maximum marks: {max_marks}")
print(f"Minimum marks: {min_marks}")

#find the best performing student
best_student = np.argmax(avg_marks_per_student)
print(f"\nBest Performing Student: Student{best_student + 1}")
print(f"Their average marks: {avg_marks_per_student[best_student]: .2f}")
```
