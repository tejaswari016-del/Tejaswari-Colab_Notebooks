```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error,mean_squared_error

#random data which entered in excel sheet

temperature = np.array([20,19,18,19,18,19,18,17,18,17,16,17,16,17,16,17,16,17,16,15])
actual = np.array([27.1,28.4,30.2,30,31.8,31.6,33.4,35,34.7,36.4,38,37.6,39.3,39,40.7,40.4,42.1,41.7,43.4,45.1])

rainfall = np.array([760,780,800,820,840,860,880,900,920,940,960,980,1000,1020,1040,1060,1080,1100,1120,1140])

#Regression equation
predicted = 2.05+0.0398*rainfall-0.298*temperature

#Create dataFrame
df = pd.DataFrame({
    "Rainfall":rainfall,
    "Temperature":temperature,
    "Actual":actual,
    "Predicted":predicted
})

#Absolute Error
df["Absolute Error"] = abs(df["Actual"]-df["Predicted"])

#Squared error
df["Squared Error"] = (df["Actual"]-df["Predicted"])**2

#Evaluation metrics
mae = mean_absolute_error(actual,predicted)
mse = mean_squared_error(actual,predicted)
rmse = np.sqrt(mse)

print(df)

print("\nModel Evaluation Metrics")
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```
