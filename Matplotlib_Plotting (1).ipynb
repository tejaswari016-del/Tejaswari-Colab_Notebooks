```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,25,20,30,40]
plt.plot(x,y, marker = '*',linestyle = '--',color = 'blue',linewidth = 2,markersize = 8,label = 'Sales')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
sizes = 1000*np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.6)
plt.colorbar()
plt.title("Advanced Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
#Example: Crop yield over years
import matplotlib.pyplot as plt
years = [2019,2020,2021,2022,2023]
yield_kg = [2500,2700,3000,3200,3500]

plt.plot(
    years,yield_kg,
    color='red',
    marker='X',
    linestyle='--',
    linewidth=2,
    markersize=10,
    label='Crop yield'
  )
plt.title("CROP YIELD OVER YEARS")
plt.xlabel("Year")
plt.ylabel("Yield(kg/ha)")
plt.legend()
plt.grid(True)
plt.show()
```
