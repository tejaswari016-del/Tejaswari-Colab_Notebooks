```python
def load_data(filename):
  crops = []
  file = open(filename, "r")
  for line in file:
    line = line.strip()
    if line:
      data = line.split(",")
      crop = {
          "name":data[0],
          "season":data[1],
          "area":float(data[2]),
          "yield":float(data[3]),
          "cost":float(data[4]),
          "revenue":float(data[5])
      }
      crops.append(crop)
  file.close()
  return crops

def display_crops(crops):
  """To display all the crop details"""
  print("\nCROP DETAILS: ")
  for crop in crops:
      print(crop)

def find_kharif_crops(crops):
  """TO find and display all the kharif crops"""
  print("\nKHARIF CROPS:")
  for crop in crops:
         if crop["season"]=="kharif":
            print(crop["name"])

def find_non_kharif_crops(crops):
  print("\nNON-KHARIF CROPS: ")
  for crop in crops:
         if crop["season"]=="rabi":
            print(crop["name"])

def calculate_profit(crops):
  print("\nCALCULATED PROFIT:")
  for crop in crops:
      profit = crop["revenue"]-crop["cost"]
      print(crop["name"],"Profit:",profit)


def summary_stats(crops):
  print("\nSUMMARY STATISTICS: ")
  total_area = 0
  total_yield = 0

  for crop in crops:
     total_area += crop["area"]
     total_yield += crop["yield"]

  avg_yield = total_yield / len(crops)

  print("Total area: ",total_area)
  print("Average yield: ",avg_yield)

#-----------Main program----------
def main():
    input_file = "profit.analysis.txt"
    output_file = "analysis_report.txt"

    crops = load_data(input_file)

    for crop in crops:
      print(crop)
    for crop in crops:
     for key,value in crop.items():
      print(f"{key}: {value}")
    print("------------")

    display_crops(crops)
    find_kharif_crops(crops)
    find_non_kharif_crops(crops)
    calculate_profit(crops)
    max_min_profit_crop(crops)
    summary_stats(crops)

main()

```
